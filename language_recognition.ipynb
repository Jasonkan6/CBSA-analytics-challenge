```python
# Function to identify the language of a news article
def identify_language(text):
    try:
      lang = langid.classify(text)[0]
    except:
      print("error when identifying language: ", text)
      lang = ['error']
    return lang
```

```python
## Apply language identification to all news articles
def rec(start, end):
  df["language"][start:end] = df['content'][start:end].apply(identify_language)
```

```python
df['language'] = None
start = 0
end = 10000
for i in range(start, end):
  print("start: ", start, ", end: ", end)
  rec(start, end)
  start += 10000
  end += 10000
  if (end>=df.shape[0]):
    end = df.shape[0]
```

```python
## Number of different languages
print(df.language.value_counts())
```

```python
## News that are not identified as in english/chinese
specific_value_rows = df[~df['language'].isin(['en', 'zh'])]
print(specific_value_rows.content)
```
