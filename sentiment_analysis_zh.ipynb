```python
zh_df = pd.read_csv('sent_ana/zh.csv')
```

```python
zh_df = zh_df.drop('docid', axis=1)
```

```python
zh_df['content'] = zh_df['content'].fillna("")
```

```python
zh_df['pubdate'] = pd.to_datetime(zh_df['pubdate'])
# print(zh_df)
```

```python
df_sorted = zh_df.sort_values('pubdate')
# print(df_sorted)
```

```python
df_sorted.set_index('pubdate', inplace=True)
# Group the data into weekly groups
weekly_groups = df_sorted.resample('W')
group_names = []
# Iterate over the weekly groups and perform operations
for group_name, group_data in weekly_groups:
    group_names.append(group_name)
    # print("Group:", group_name) # group_name indicates the end of the week
    # print(len(group_data))
    # print(group_data.content)
    # print("--------------------")
```

```python
# print(group_names)
```

```python
def analyze_sentiment(text):
    s = SnowNLP(text)
    return s.sentiments
```

```python
from tqdm import tqdm
progress_bar = tqdm(group_names[80:81], desc='Processing Groups')

for g in progress_bar:
  tone = weekly_groups.get_group(g)['content'].apply(lambda x: analyze_sentiment(x))
  tone.to_frame()['content'].to_csv('sent_ana/tone'+g.strftime("%Y-%m-%d %H:%M:%S")+'.csv', index=False, header=False)
```
